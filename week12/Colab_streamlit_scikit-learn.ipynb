```python
%%writefile sckit.py

import streamlit as st
from sklearn import datasets
import pandas as pd

load_datasets = [name for name in datasets.__all__ if name.startswith("load")]

# Create a Streamlit app
st.title("Load Datasets")

# Create a selectbox to choose a dataset
load_datasets = [name for name in datasets.__all__ if name.startswith("load")]
selected_dataset = st.selectbox("Choose a dataset", load_datasets)
st.write(f"You selected: {selected_dataset}")

# Load the selected dataset
if st.button("Load"):
  dataset = getattr(datasets, selected_dataset)()
  df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
  df['target'] = dataset.target

  # Print the dataset information
  st.write(getattr(datasets, load_datasets[0])().keys())
  st.write(dataset.DESCR)
  st.write(df.head(10))

# 주성분 분석
st.title("Principal Component Analysis Visualization")
import sklearn.decomposition as decomposition
import matplotlib.pyplot as plt
import seaborn as sns

# Reduce dimensionality to 2 principal components
pca = decomposition.PCA(n_components=2)
pca.fit(dataset.data)
df_pca = pd.DataFrame(pca.transform(dataset.data), columns=['PC1', 'PC2'])
df_pca['target'] = dataset.target

# Create the scatter plot
fig, ax = plt.subplots()
scatter_plot = sns.scatterplot(data=df_pca, x='PC1', y='PC2',
                               hue='target', palette='viridis', legend='full',
                               ax=ax)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title(f'PCA of Iris Dataset from the{selected_dataset} dataset')
# Display the plot in the Streamlit app
st.pyplot(fig)
```

```python
# Start the Streamlit app in the background

from pyngrok import ngrok
import os

# Replace 'YOUR_NGROK_AUTH_TOKEN' with your actual ngrok AuthToken
NGROK_AUTH_TOKEN = '?????????????????????????'
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

def run():
    filename = input('Enter your python file name to run')
    os.system(f'streamlit run {filename} &')
    # Create a public URL using ngrok
    http_tunnel = ngrok.connect(8501, "http")
    print(f"Streamlit app is live at: {http_tunnel.public_url}")

run()
```
