```python
import pandas as pd
from sqlalchemy import create_engine
import schedule
import time

# 데이터베이스 연결 설정 (예: SQLite)
DATABASE_URI = 'sqlite:///example.db'

# ETL 작업 함수 정의
def etl_job():
    print("Starting ETL job...")

    # Extract: CSV 파일에서 데이터 읽기
    url = "https://kanaries-app.s3.ap-northeast-1.amazonaws.com/public-datasets/bike_sharing_dc.csv"
    df = pd.read_csv(url, parse_dates=['date'])

    # Transform: 필요한 데이터 변환 수행 (예: 데이터 필터링)
    df_filtered = df[df['count'] > 100]  # count가 100 이상인 데이터만 필터링

    # Load: 데이터베이스에 저장
    engine = create_engine(DATABASE_URI)
    df_filtered.to_sql('bike_sharing', engine, if_exists='replace', index=False)

    print("ETL job completed.")

# 스케줄링 설정 (예: 매일 오전 9시에 실행)
schedule.every().day.at("09:00").do(etl_job)
print("Scheduler started. Waiting for scheduled tasks...")
```

```python
etl_job()
```

```python

# 스케줄러 실행
try:
    while True:
        schedule.run_pending()
        time.sleep(1)
except KeyboardInterrupt:
    print("Scheduler stopped.")
```
