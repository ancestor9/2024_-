## **Step 3: Mage.AI 서버 시작**
- Mage.AI 서버를 시작:  Colab 환경에서는 로컬 서버를 시작한 후, ngrok을 사용하여 외부에서 접근이 가능함

```python
# Ngrok 설치
# !pip install pyngrok

# Mage.AI 서버 시작
import subprocess
import threading
from pyngrok import ngrok

# Ngrok 터널 생성
ngrok.set_auth_token('2Tc747eKbLpLOKXbVplhOqcgPxt_5u8uUMJo95CSoXeGvLJn5')  # Ngrok 계정의 인증 토큰 사용
public_url = ngrok.connect(6789)

# Mage.AI 서버 시작 함수
def start_mage():
    subprocess.run(['mage', 'start', 'my_mage_project'])

# 서버 시작을 위한 스레드 생성
thread = threading.Thread(target=start_mage)
thread.start()

# Ngrok URL 출력
print(f'Mage.AI UI is accessible at: {public_url}')
```
