### **다른 방법 corpus**

```python
corpus = []
for keywords in df['특성추출(가중치순 상위 50개)']:
    corpus.append(keywords)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer,TfidfTransformer
import numpy as np

stop_words = ['no']
vect = CountVectorizer(stop_words = stop_words, ngram_range=(1, 1), min_df= 0.005, max_df= 0.8)
X = vect.fit_transform(corpus) # Doc-Term Matrix 계산
X = TfidfTransformer().fit_transform(X)

count = X.toarray().sum(axis=0) # Doc-Term Matrix 각 단어에 대한 합
idx = np.argsort(-count)
count = count[idx]

feature_name = np.array(vect.get_feature_names_out())[idx]
keyword_list = list(zip(feature_name, count))
result_df = pd.DataFrame(keyword_list)
```

```python
result_df
```

## **사용자정의함수로 실습**

```python
# 사용자 정의함수를 이용하여 만들어 보기

def text_simulator() :
    text = [np.random.choice(['apple', 'banana', 'melon', 'orange']) for i in range(1000)]
    mydict = dict()

    for key in text:
        if key in mydict :
            mydict[key] += 1
        else :
            mydict[key] = 1

    return mydict
```

```python
text_simulator()
```

```python
# 100회 실행하기
your_dict = []
for _ in range(1000):
    result = text_simulator()
    your_dict.append(result)
```

```python
your_dict[:5]
```

```python
import pandas as pd
df = pd.DataFrame(your_dict)
df.head()
```
