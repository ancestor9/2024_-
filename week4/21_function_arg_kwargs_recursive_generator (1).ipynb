### **여러 명을 불러 인사할 때**

```python
def greet(*name, greeting="Hello"):
    print(f"{greeting}, {name}!")
```

```python
# 일반 인자
greet("Carol", "Sanggoo")  # 출력: Hello, ('Carol', 'Sanggoo')!
```

```python
# 일반 인자
greet("Carol", "Sanggoo", "Breece")  # 출력: Hello, ('Carol', 'Sanggoo', 'Breece')!
```

### **여러 명을 불러 인사하고 안녕할 때**

```python
def greet(*name, **kwargs):
    print(f"{kwargs}, {name}!")
```

```python
greet("Carol", "Sanggoo", "Breece", greet1='Hi', greet2='Bye')
```

>>> **일반인자(arguments)**

>>> **사전형 키워드 인자(keywords arguments)**


- List로 부르면 안된다.

```python
def f(**kwargs):
    print(type(kwargs))
    kwargs = [kwarg for arg in kwargs]
    print(f'U type kwargs : {kwargs}')
```

```python
f(10,20,30)
```

```python
f(a=10, b=20, c=30)
```

```python
def f(**kwargs):
    print(type(kwargs))  # This will print "<class 'dict'>"
    for key, value in kwargs.items():
        print(f'Key: {key}, Value: {value}')
```

```python
f(b=20, c=30)
```

```python
def f(**kwargs):
    print(type(kwargs))
    print(f'U type kwargs : {kwargs}')
```

```python
f(a=10, b=20, c=30)
```

```python
def f(**kwargs):
    print(type(kwargs))
    for key, value in kwargs.items():
        print(f'key: {key}, value: {value}')
    print(f'U type kwargs : {kwargs}')
```

```python
f(a=10, b=20, c=30)
```

>> ##### 1. *args 형식을 사용하면 모든 형식의 인자를 받을 수 있다(Tuple).
>> ##### 2. *kwarg 형식을 사용하면 key와 value형식의 인자를 받을 수 있다(Dictionary).
>> ##### 3. *args, *kwarg 형식을 같이 사용하면 모든 형식의 인자를 받을 수 있다(Dictionary).

## **재귀함수**

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# 팩토리얼을 계산하려면 양수 정수를 인수로 전달합니다.
result = factorial(5)
print(f'5의 팩토리얼은 {result}입니다.')  # 출력: "5의 팩토리얼은 120입니다."
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
```

```python
[fibonacci(i) for i in range(11)]
```

## **Generator**

### **피노나치 Generator 만들기**

```python
def fib_generator(n):
    curr, next = 0, 1
    for _ in range(n):
        yield curr
        curr, next = next,  curr + next
```

```python
fib = fib_generator(n=11)
fib
```

```python
fib
```

```python
for i, val in enumerate(fib):
    print(f'Fibonacci({i}): {val}')
```

```python
for i, val in enumerate(fib):
    print(f'Fibonacci({i}): {val}')
```

### **피노나치 반복문과 Generator 비교으로 만들기**

```python
def fib(n):
    curr, next = 0, 1
    for _ in range(n):
        curr, next = next,  curr + next

    return curr
```

```python
fib(n=100)
```

```python
[fibonacci(i) for i in range(15)]
```

```python
def fib_generator(n):
    curr, next = 0, 1
    for _ in range(n):
        yield curr
        curr, next = next,  curr + next
```

```python
from itertools import islice
# 제너레이터 생성
gen = fib_generator(n=15)

# 처음 5개 아이템만 가져오기
first_5_items = list(islice(gen, 5))
print(first_5_items)  # 출력: [1, 2, 3, 4, 5]

# 나머지 아이템 가져오기
remaining_items = list(gen)
print(remaining_items)  # 출력: [6, 7, 8, 9, 10]
```
